```python
import json
import pandas as pd
from datetime import datetime, timedelta
```

```python
data = json.load(open('url_patterns.json'))

# create a list of dates from 2011-09-01 to 2022-11-22
start_date = datetime.strptime('2011-09-01', '%Y-%m-%d')
end_date = datetime.strptime('2022-11-22', '%Y-%m-%d')
datex = {(start_date + timedelta(days=x)).strftime('%Y-%m-%d')  : 0 for x in range(0, (end_date-start_date).days)}

years = []
months = []
dates = []
patterns = []

different_lens = set()
for year in data:
    for month in data[year]:
        for date in data[year][month]:
            for pattern in data[year][month][date]:
                if len(pattern[0]) == 1:
                    pattern[0] = pattern[0][0]
                if len(pattern[1]) == 1:
                    pattern[1] = pattern[1][0]
            years.append(year)
            months.append(month)
            dates.append(date)
            patterns.append(pattern)
            if(date == '00'):
                # Remove all dates in a month
                for i in range(1, 32):
                    try:
                        del datex[f'{year}-{month}-{i:02d}']
                    except:
                        pass
            else:
                try:
                    del datex[f'{year}-{month}-{date}']
                except:
                    pass
            if(len(data[year][month][date])) == 1:
                data[year][month][date] = data[year][month][date][0]
            else:
                print(year, month, date, data[year][month][date])


df = pd.DataFrame({'year': years, 'month': months, 'date': dates, 'pattern': patterns})
df.to_csv('url_patterns.csv', index=False)

data = json.dumps(data)
with open('url_patterns_compressed.json', 'w') as f:
    f.write(data)
```
